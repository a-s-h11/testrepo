```python
#note: the provided dataset with the project is the same as the AmesHousing dataset

#step1: import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#step2: load the given dataset
df = pd.read_csv('AmesHousing.csv')

#step3: data preprocessing (equate na values to 0)
df = df.fillna(0)

#step4: one-hot encoding (convert categorical columns to numerical columns)
cat_col = df.select_dtypes(include = ['object']).columns

#step5: merge using fill_dummies and drop unnecessary columns
df_final = pd.get_dummies(df, columns = cat_col, drop_first = True)

#step6: train your model
model = LinearRegression()
x = df_final.drop('SalePrice', axis = 'columns') #equates to df_final without SalePrice
y = df_final.SalePrice #equates to SalePrice
model.fit(x, y)

#step7: predict SalePrice
y_predict = model.predict(x)

#step8: plot actual sale price vs predicted sale price graph
plt.scatter(y, y_predict, alpha = 0.25) #actual values on x axis, predicted values on y axis
plt.xlabel('actual sale price')
plt.ylabel('predicted sale price')
plt.title('actual sale price vs predicted sale price')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color = 'black') #generates a line where y_predict = y (predicted sale price = actual sale price)
plt.show()
```
